```python
# 1. 正确导入
from langchain_community.llms import Tongyi
import os
from dotenv import load_dotenv

# 2. 加载环境变量
load_dotenv()
api_key = os.getenv("DASHSCOPE_API_KEY")

# 3. 初始化模型（注意 model 参数是字符串）
llm = Tongyi(
    model="qwen-turbo",  # 或者用 qwen-plus / qwen-max
    api_key=api_key,
    temperature=0.1
)

# 4. 调用并打印
response = llm.invoke("夏天适合吃什么水果？回复几个水果名称就可以了！")
print(response)
```

```python
# 智谱AI
import os
from dotenv import load_dotenv
from langchain_community.chat_models import ChatZhipuAI
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage
load_dotenv()
chat = ChatZhipuAI(
    model="glm-4",
    temperature=0.5,# 模型温度，值一般在0到1之间，值越小随机性越低（也就是模型的发散思维，值越大发散思维越强）
    zhipuai_api_key = os.getenv("ZHIPUAI_API_KEY")
)

messages = [
    AIMessage(content="Hi."),#初始化时的信息（可有可无）
    SystemMessage(content="你是一个助手，请用中文回复"),#模型扮演的角色
    HumanMessage(content="夏天适合吃什么水果？"),#人类的问题
]

response = chat.invoke(messages)
print(response.content)
```
